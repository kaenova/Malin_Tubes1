```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("/content/drive/MyDrive/Semester 5/Pembelajaran Mesin/Tugas Besar/Dataset/Processed/kendaraan_train_explored_balance.csv")
```

```python
df['Tertarik'].unique()
```

```python
labels = np.array(df['Tertarik'].to_list())
```

```python
df_no_labels = df.drop(columns="Tertarik", axis=1)
```

```python
df_no_ids = df_no_labels.drop(columns="id", axis=1)
```

```python
df_no_ids
```

```python
training_data = df_no_ids.to_numpy()
```

# K - Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2).fit(training_data)
```

```python
prediction = kmeans.predict(training_data)
```

```python
import sklearn
```

```python
sklearn.metrics.accuracy_score(labels, prediction)
```

# K - Medoids (not Good for big Data >30.000 x 10)

```python
input = training_data[:30000]
output = labels[:30000]
```

```python
from sklearn_extra.cluster import KMedoids
```

```python
kmedoids = KMedoids(n_clusters=2, max_iter=1000).fit(input)
```

```python
prediction = kmedoids.predict(input)
```

```python
sklearn.metrics.accuracy_score(output, prediction)
```

# Agglomerative Clustering (not Good for Big Data)

```python
input = training_data[:30000]
output = labels[:30000]
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
prediction = AgglomerativeClustering(n_clusters=2).fit_predict(input, output)
```

```python
sklearn.metrics.accuracy_score(output, prediction)
```

# Birch (no good big data)

```python
input = training_data[:30000]
output = labels[:30000]
```

```python
from sklearn.cluster import Birch
```

```python
prediction = Birch(n_clusters=2).fit_predict(input, output)
sklearn.metrics.accuracy_score(output, prediction)
```

# Spectral Clustering

```python
input = training_data[:5000]
output = labels[:5000]
```

```python
from sklearn.cluster import SpectralClustering
```

```python
prediction = SpectralClustering(n_clusters=2).fit_predict(input, output)
sklearn.metrics.accuracy_score(output, prediction)
```

# Gaussian Mixture

```python
from sklearn.mixture import GaussianMixture
```

```python
prediction = GaussianMixture(n_components=2).fit_predict(input, output)
sklearn.metrics.accuracy_score(output, prediction)
```
